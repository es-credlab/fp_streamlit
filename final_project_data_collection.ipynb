Get QID's of Feminists' Wikipedia Articles using Enlgish Titles -

```python
pip install bs4
```

```python
import requests
from bs4 import BeautifulSoup as BS
```

```python
list_of_feminists = pd.read_csv("data\list_of_feminists.csv")
```

```python
url = "https://en.wikipedia.org/wiki/List_of_feminists"
headers = {'User-Agent': 'A simple bot for my Python class (myname@website.info)'}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    domTree = BS(response.text, 'html.parser')
    fullPage = domTree.find('body')
else:
    print("Url Request Failed")
```

```python
print(fullPage)
```

```python
len(list_of_feminists["Name"])
```

```python
titles = []

for name in list_of_feminists["Name"]:
    article_info = fullPage.find_all('td', attrs={'data-sort-value': name})
    if article_info == []:
        article_info = fullPage.find_all('td', attrs={'data-sort-value': f"{name.split()[-1]}, {' '.join(name.split()[:-1])}"})
    if article_info == []:
        article_info = fullPage.find_all('td', attrs={'data-sort-value': f"{' '.join(name.split()[1:])}, {name.split()[0]}"})
    if article_info == []:
        article_info = fullPage.find_all('td', attrs={'data-sort-value': f"{name.split()[0]}"})
    if article_info != []:
        tag = article_info[0]
        titles.append(tag.find('a')['title'])
    else:
        titles.append(None)
                     
titles
```

```python
len(titles)
```

```python
site = mwclient.Site('en.wikipedia.org')

result = site.api(
                'query',
                prop='pageprops',
                titles='Helen of Anjou, Queen of Serbia',
                pprop='wikibase_item'
            )

result
```

```python
site = mwclient.Site('en.wikipedia.org')

q_id_list = []

index = 0

for title in titles:
    try:    
        result = site.api(
                'query',
                prop='pageprops',
                titles=title,
                pprop='wikibase_item'
            )

        unique_key = list((result['query']['pages']).keys())[0]

        q_id  = (result['query']['pages'][unique_key]['pageprops']['wikibase_item'])
            
        q_id_list.append(q_id)
    
    except:
        try:
            print("Error: Didn't work for some reason - ")
        except:
            print("No pageprops for some reason?")
        q_id_list.append(None)
    
    # print(index) # making sure it's not getting stuck anywhere and tracking progress
    # index+=1 
```

Assigning + Cleaning Dates in Pageviews CSV

```python
date_range = pd.date_range(start='2024-01-01', end='2024-01-07')
```

```python
type(date_range)
```

```python
cleaned_pageview_dates = pageview_data[pageview_data['date_object'].isin(date_range)]
```

```python
cleaned_pageview_dates.head()
```

```python
len(cleaned_pageview_dates)
```

```python
cleaned_pageview_dates = cleaned_pageview_dates[['date','qid','pageviews','date_object','wave']]
```

```python
final_pageview_data = pd.DataFrame(columns=['date','qid','pageviews','date_object','wave'])
```

```python
len(date_range)
```

```python
len(pd.DataFrame({'date':date_range,'qid':[q_id]*7,'pageviews':[0]*7,'date_object':date_range}))
```

```python
all_qids = set(list_of_feminists['q_id'])
data_qids = set(pageview_data['qid'].drop_duplicates())

for q_id in all_qids:
    if q_id not in data_qids:
        cleaned_pageview_dates = pd.concat([cleaned_pageview_dates,pd.DataFrame({'date':date_range,'qid':[q_id]*7,'pageviews':[0]*7,'date_object':date_range})])
    else:
        qid_info = pageview_data[pageview_data['qid'] == q_id]
        all_dates = {'date':date_range,'qid':[q_id]*7,'pageviews':[],'date_object':date_range,'wave':[list(qid_info['wave'])[0]]*7}
        for date in date_range:
            if date not in list(qid_info['date_object']):
                all_dates['pageviews'].append(0)
            else:
                all_dates['pageviews'].append(int(qid_info[qid_info['date_object'] == date]['pageviews'].values[0]))
        cleaned_pageview_dates = pd.concat([cleaned_pageview_dates,pd.DataFrame(all_dates)])
```

Assigning Waves of Feminism in Pageviews CSV

```python
cleaned_qids = list(cleaned_list_of_feminists['q_id'])
cleaned_waves = list(cleaned_list_of_feminists['wave'])
```

```python
new_wave_data = []

for qid in pageview_data['qid']:
    if qid in cleaned_qids:
        index = cleaned_qids.index(qid)
        wave = cleaned_waves[index]
        new_wave_data.append(wave)
    else:
        new_wave_data.append(None)

pageview_data['wave'] = new_wave_data
```
